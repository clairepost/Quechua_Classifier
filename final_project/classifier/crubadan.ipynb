```python
import nltk
nltk.download('crubadan')
```

```python
from nltk.corpus import crubadan

# List all supported languages
print(crubadan.langs())
```

```python
# Access ngram frequency distributions for 'quh' and 'qug'
quh_fd = crubadan.lang_freq('quh')
qug_fd = crubadan.lang_freq('qug')
```

```python
# Example: Print the frequency of a specific ngram in both dialects
ngram_example = 'mam'  # You can replace this with any ngram of interest
print(f"Frequency of '{ngram_example}' in 'quh': {quh_fd[ngram_example]}")
print(f"Frequency of '{ngram_example}' in 'qug': {qug_fd[ngram_example]}")
```

```python
# Compare top 10 most common ngrams in both dialects
print("Top 10 ngrams in 'quh':")
for ngram, freq in quh_fd.most_common(10):
    print(f"{ngram}: {freq}")

print("\nTop 10 ngrams in 'qug':")
for ngram, freq in qug_fd.most_common(10):
    print(f"{ngram}: {freq}")
```

```python
import matplotlib.pyplot as plt

# Create a bar chart for visualization
def plot_ngram_frequencies(freq_dist, title):
    ngrams = [ngram for ngram, _ in freq_dist.most_common(10)]
    frequencies = [freq for _, freq in freq_dist.most_common(10)]
    plt.figure(figsize=(10, 5))
    plt.bar(ngrams, frequencies, color='blue')
    plt.xlabel('Ngrams')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.show()

plot_ngram_frequencies(quh_fd, "Top 10 N-grams in Quechua Dialect 'QUH'")
plot_ngram_frequencies(qug_fd, "Top 10 N-grams in Quechua Dialect 'QUG'")
```
